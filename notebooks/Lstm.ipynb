```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['Text'], data['Label'], test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import math


class DatasetGenerator(keras.utils.Sequence):
    def __init__(self, X, y, num_features:int, batch_size: int = 16):
        self.x, self.y = X, y
        self.batch_size = batch_size
        self.num_features = num_features
    
    # return number of batches
    def __len__(self):
        return int(np.ceil(len(self.x) / self.batch_size))
    
    # return a single batch as tuple (inputs, ouput) 
    def __getitem__(self, idx):
        batch_x = self.x[idx * self.batch_size : (idx + 1) * self.batch_size]
        batch_y = self.y[idx * self.batch_size : (idx + 1) * self.batch_size]

        return keras.utils.to_categorical(y=batch_x, num_classes=self.num_features), batch_y

# Initialize and fit the tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train.astype(str).tolist())
vocab_size = len(tokenizer.word_index) + 1

# integer encode the documents
encoded_docs = tokenizer.texts_to_sequences(X_train.tolist())
X_test = np.array(tokenizer.texts_to_sequences(X_test.tolist()))
print('encoded_docs:\n',encoded_docs)

# pad documents to a max length of 4 words
max_length = 90

X_train = pad_sequences(encoded_docs, maxlen=max_length, 
padding='pre',truncating="pre",value=0)
X_test = pad_sequences(
    sequences=X_test,
    maxlen=int(max_length), 
    padding="pre", 
    truncating="pre", 
    value=0
)

train_dataset = DatasetGenerator(X=X_train, y=pd.get_dummies(y_train).to_numpy(), num_features=vocab_size, batch_size=64)
test_dataset = DatasetGenerator(X=X_test, y=pd.get_dummies(y_test).to_numpy(), num_features=vocab_size, batch_size=64)
```

```python
import gc
del encoded_docs
gc.collect()
```

```python
model= keras.Sequential([
keras.Input(shape=(90,vocab_size)),
      keras.layers.GRU(units=16, dropout=0.2, recurrent_dropout=0.2 ,return_sequences=False),
      keras.layers.Dense(units=4,activation="softmax")
      ])
```

```python

keras_callbacks   = [
    keras.callbacks.ModelCheckpoint(filepath = 'best_model.h5', monitor = 'val_loss',
                              save_best_only = True, mode = 'auto', verbose = 1),
]
```

```python
# compile the model
optimizer = keras.optimizers.Nadam()
model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics='CategoricalAccuracy')
# summarize the model
print(model.summary())
```

```python
import pickle

# saving
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
history= model.fit(train_dataset,validation_data=test_dataset,batch_size=64, epochs=4,callbacks=[keras_callbacks])
```

```python
import time
from sklearn.metrics import classification_report
```

```python
start= time.time()
y_pred= model.predict(test_dataset)
end= time.time()
print("Prediction time = ", end-start)
print(classification_report(y_pred.argmax(axis= 1),y_test.tolist()))
```
